## Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.

### Answer:
Deep Learning is a subfield of Machine Learning that uses artificial neural networks with multiple hidden layers to automatically learn complex patterns from large amounts of data. It is inspired by the structure and functioning of the human brain.

Deep learning is especially effective for tasks such as image recognition, speech recognition, natural language processing, and autonomous systems.

---

### Evolution of Deep Learning:
1. **Traditional Programming**  
   Humans explicitly define rules and logic to solve problems.

2. **Machine Learning**  
   Models learn patterns from data using manually engineered features.

3. **Artificial Neural Networks**  
   Introduction of neurons, weights, bias, and activation functions.

4. **Deep Learning**  
   Neural networks with many hidden layers capable of learning hierarchical and abstract features automatically.

---

### Difference Between Machine Learning and Deep Learning:

| Machine Learning | Deep Learning |
|------------------|---------------|
| Requires manual feature extraction | Automatically learns features |
| Performs well on structured data | Performs well on unstructured data |
| Less computational power needed | Requires high computational power (GPUs) |
| Limited performance on complex tasks | High performance on complex tasks |

---

### Example:
In image classification, Machine Learning requires manual features such as edges and textures, whereas Deep Learning learns features directly from raw images.

---

### Conclusion:
Deep Learning represents an advanced stage of Machine Learning, enabling models to solve highly complex problems with minimal human intervention in feature design.


## Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?

### Answer:
A perceptron is the simplest form of an artificial neural network. It is a single-layer neural network mainly used for binary classification problems.

---

### Basic Architecture of a Perceptron:
The perceptron consists of the following components:
- **Input Layer:** Receives input features
- **Weights:** Assign importance to each input
- **Bias:** Shifts the decision boundary
- **Summation Unit:** Computes weighted sum of inputs
- **Activation Function:** Produces the final output

---

### Mathematical Representation:
\[
y = f\left(\sum_{i=1}^{n} w_i x_i + b\right)
\]

where  
\(x_i\) = input features  
\(w_i\) = weights  
\(b\) = bias  
\(f\) = activation function  

---

### Functioning of a Perceptron:
1. Inputs are multiplied by corresponding weights.
2. The weighted inputs are summed.
3. Bias is added to the sum.
4. The activation function determines the output (usually 0 or 1).

---

### Limitations of a Perceptron:
1. **Cannot Solve Non-Linear Problems**  
   It fails to classify problems like XOR.

2. **Single Layer Only**  
   No hidden layers, limiting learning capacity.

3. **Binary Output Only**  
   Works only for binary classification tasks.

4. **Linearly Separable Data Requirement**  
   Performs well only when data is linearly separable.

---

### Conclusion:
While the perceptron is fundamental to understanding neural networks, its limitations led to the development of multi-layer neural networks and deep learning models.

## Question 3:Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

### Answer:
An activation function is a mathematical function applied to the weighted sum of inputs in a neural network neuron. It determines whether a neuron should be activated or not and what output it should produce.

Activation functions are essential because they introduce **non-linearity** into the network, enabling it to learn complex patterns and relationships in data.

---

### Purpose of Activation Functions:
1. **Introduce Non-Linearity**  
   Without activation functions, neural networks behave like linear models and cannot learn complex relationships.

2. **Control Neuron Output**  
   Activation functions decide the range and form of the output values.

3. **Enable Deep Learning**  
   They allow stacking of multiple layers to solve complex problems.

4. **Improve Learning Capability**  
   Help the network approximate complex functions.

---

### Comparison of Sigmoid, ReLU, and Tanh Activation Functions:

| Activation Function | Formula | Output Range | Advantages | Disadvantages |
|--------------------|---------|--------------|------------|---------------|
| **Sigmoid** | \( \frac{1}{1 + e^{-x}} \) | (0, 1) | Smooth, probabilistic output | Vanishing gradient, slow training |
| **ReLU** | \( \max(0, x) \) | [0, ∞) | Fast, efficient, avoids vanishing gradient | Dying ReLU problem |
| **Tanh** | \( \frac{e^x - e^{-x}}{e^x + e^{-x}} \) | (-1, 1) | Zero-centered output, better than sigmoid | Still suffers from vanishing gradient |

---

### Use Cases:
- **Sigmoid:** Output layer for binary classification
- **ReLU:** Hidden layers in deep neural networks
- **Tanh:** Hidden layers when zero-centered output is preferred

---

### Conclusion:
Activation functions play a vital role in neural networks by enabling non-linear learning. Among the three, **ReLU** is most commonly used in hidden layers due to its efficiency, while **Sigmoid** and **Tanh** are useful in specific output and hidden layer scenarios.

## Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.

### Answer:

In neural networks, loss function and cost function are used to measure how well a model is performing. Although they are closely related, they differ in scope and usage.

---

### Loss Function:
A loss function measures the error between the predicted output and the actual output **for a single training example**. It quantifies how wrong the model’s prediction is for one data point.

#### Characteristics:
- Calculated for one data sample
- Used during backpropagation
- Guides weight updates

#### Examples of Loss Functions:
1. **Mean Squared Error (MSE) Loss**
\[
L(y, \hat{y}) = (y - \hat{y})^2
\]

2. **Binary Cross-Entropy Loss**
\[
L = -[y \log(\hat{y}) + (1 - y)\log(1 - \hat{y})]
\]

3. **Categorical Cross-Entropy Loss**
Used for multi-class classification problems.

---

### Cost Function:
A cost function represents the **average loss over the entire training dataset**. It provides a single value that summarizes the model’s overall performance.

#### Characteristics:
- Calculated over all training samples
- Used to evaluate overall model performance
- Optimizers aim to minimize the cost function

#### Example of Cost Function:
If MSE is used as the loss function, the cost function becomes:
\[
J(w) = \frac{1}{N}\sum_{i=1}^{N}(y_i - \hat{y}_i)^2
\]

---

### Key Differences Between Loss Function and Cost Function:

| Loss Function | Cost Function |
|--------------|---------------|
| Measures error for one data point | Measures average error for all data points |
| Used during backpropagation | Used to evaluate overall training |
| Helps update weights | Helps compare models |
| Sample-level | Dataset-level |

---

### Example Explanation:
In a neural network predicting house prices:
- **Loss Function:** Error for one house prediction
- **Cost Function:** Average error across all houses in the dataset

---

### Conclusion:
The loss function evaluates individual prediction errors, while the cost function aggregates these errors over the dataset. Optimizers minimize the cost function by reducing individual losses, enabling the neural network to learn effectively.

---




## Question 5:What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.

### Answer:

Optimizers are algorithms used in neural networks to update the model’s weights and biases during training. Their main goal is to minimize the loss function by efficiently adjusting parameters using gradient information obtained through backpropagation.

Without optimizers, a neural network cannot learn effectively because weights would not be updated in the correct direction or at the correct speed.

---

### Role of Optimizers in Neural Networks:
1. **Minimize Loss Function**  
   Optimizers reduce the error between predicted output and actual output.

2. **Update Weights and Biases**  
   They adjust parameters in the direction that reduces loss.

3. **Improve Convergence Speed**  
   Optimizers help the model reach optimal solutions faster.

4. **Handle Large and Complex Data**  
   Advanced optimizers adapt learning rates for better stability.

5. **Avoid Poor Local Minima**  
   Some optimizers use momentum to escape shallow local minima.

---

### 1. Gradient Descent
Gradient Descent is the most basic optimization algorithm. It updates model parameters using the gradient of the loss function calculated over the entire training dataset.

**Weight Update Rule:**
\[
w = w - \eta \frac{\partial L}{\partial w}
\]

**Advantages:**
- Simple and easy to understand
- Stable convergence

**Disadvantages:**
- Slow for large datasets
- Sensitive to learning rate
- Computationally expensive

---

### 2. RMSprop (Root Mean Square Propagation)
RMSprop improves Gradient Descent by using adaptive learning rates for each parameter. It divides the learning rate by the moving average of squared gradients.

**Advantages:**
- Faster convergence than Gradient Descent
- Works well with noisy data
- Reduces oscillations

**Disadvantages:**
- Requires careful tuning
- Can be unstable in some cases

---

### 3. Adam (Adaptive Moment Estimation)
Adam combines the benefits of Momentum and RMSprop. It maintains moving averages of both gradients and squared gradients, making it one of the most popular optimizers.

**Advantages:**
- Fast convergence
- Works well for large datasets
- Handles sparse gradients efficiently
- Widely used in deep learning

**Disadvantages:**
- Slightly higher memory usage
- Can sometimes overfit if not tuned properly

---

### Comparison of Optimizers:

| Optimizer | Learning Rate | Speed | Stability | Usage |
|---------|---------------|-------|-----------|-------|
| Gradient Descent | Fixed | Slow | High | Small datasets |
| RMSprop | Adaptive | Faster | Medium | Recurrent networks |
| Adam | Adaptive | Fastest | High | Deep learning models |

---

### Conclusion:
Optimizers play a crucial role in training neural networks by efficiently minimizing loss and improving learning speed. Among the three, **Adam** is the most preferred optimizer due to its fast convergence, stability, and adaptability, making it suitable for most deep learning applications.

## Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Input range
x = np.linspace(-10, 10, 500)

# NumPy implementations
def sigmoid(x):
    return 1 / (1 + np.exp(-np.clip(x, -250, 250)))  # Clip for numerical stability

def relu(x):
    return np.maximum(0, x)

def tanh_np(x):
    return np.tanh(x)


```

```python
# Compute NumPy outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh_np(x)
```

```python
# Plot activation functions
plt.figure(figsize=(12, 8))

# Sigmoid
plt.subplot(3, 1, 1)
plt.plot(x, y_sigmoid, 'b')
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("σ(x)")
plt.grid(True)

# ReLU
plt.subplot(3, 1, 2)
plt.plot(x, y_relu, 'g')
plt.title("ReLU Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid(True)

# Tanh
plt.subplot(3, 1, 3)
plt.plot(x, y_tanh, 'r')
plt.title("Tanh Function")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Question 8: Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
```

```python
(X_train, y_train), _ = mnist.load_data()
X_train = X_train / 255.0
y_train = to_categorical(y_train)
```

```python
model = Sequential([
Flatten(input_shape=(28,28)),
Dense(128, activation='relu'),
Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=5, batch_size=32)
```

```python
print("Training Accuracy:", history.history['accuracy'][-1])
```

## Question 10: You are working on a project for a bank that wants to automatically detect fraudulent transactions. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location. The goal is to classify each transaction as fraudulent or legitimate.

### Answer:
To automatically detect fraudulent transactions in a large, imbalanced dataset containing structured features such as transaction amount, merchant ID, and customer location, a well-planned machine learning pipeline is required. The goal is to accurately classify each transaction as **fraudulent** or **legitimate** while handling class imbalance and scalability.

---

### 1. Problem Type:
This is a **binary classification problem** where:
- Class 0 → Legitimate transaction  
- Class 1 → Fraudulent transaction  

---

### 2. Data Preprocessing:
- **Handle Missing Values:** Use median or mode imputation for numerical and categorical features.
- **Encoding Categorical Variables:**  
  - Use **One-Hot Encoding** or **Target Encoding** for merchant ID and location.
- **Feature Scaling:**  
  - Apply **StandardScaler** or **RobustScaler** for transaction amount to handle outliers.

---

### 3. Handling Imbalanced Data:
Since fraudulent transactions are rare, the dataset is highly imbalanced.

**Techniques to address imbalance:**
- **Resampling Methods:**  
  - Oversampling (SMOTE)  
  - Undersampling
- **Class Weights:**  
  - Assign higher weight to the fraud class.
- **Evaluation Metrics:**  
  - Precision, Recall, F1-score  
  - ROC-AUC instead of accuracy

---

### 4. Model Selection:
Suitable models for structured and imbalanced data include:
- **Logistic Regression (with class weights)**
- **Random Forest**
- **XGBoost / LightGBM (preferred)**
- **Neural Networks (MLP)**

**Tree-based ensemble models** like XGBoost perform particularly well for fraud detection.

---

### 5. Model Training:
- Split data into **training and validation sets**
- Use **cross-validation**
- Tune hyperparameters to reduce false negatives (missing fraud cases)

---

### 6. Evaluation Strategy:
Key evaluation metrics:
- **Recall:** Minimize missed fraud cases
- **Precision:** Reduce false alarms
- **F1-Score:** Balance precision and recall
- **ROC-AUC:** Overall classification performance

---

### 7. Deployment Considerations:
- **Real-time Prediction:** Fast inference is critical
- **Threshold Optimization:** Adjust decision threshold based on business cost
- **Model Monitoring:** Detect data drift and retrain periodically

---

### Conclusion:
For this fraud detection problem, a combination of **robust preprocessing**, **imbalance handling**, and **ensemble machine learning models** is ideal. Using models like XGBoost with appropriate evaluation metrics ensures high fraud detection rates while minimizing false positives, making the system effective and reliable for real-world banking applications.
