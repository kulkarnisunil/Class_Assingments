## Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.

Answer :  

- Simple Linear regession is a  supervised learning algorithm used to model the relationship between two continuous variables : one independent (predictor) variable and one dependent (response) variable.The main goal is to predict the dependent variable based on the value of the independent variable by fitting a line that best describes their relationship{understand strength and direction of relationship between two variables}.

y=mx+b

y : target varible

x : independent variable

m : slope

b : intercept

- The main purpose of SLR is to estimate how the dependent variable changes as the independent variable changes, by fitting a straight line to the observed data points.

**Purpose:==>**

1. Simple Linear Regression  finds the best-fitting straight line (regression line) through the data, which represents the average trend in the dependent variable as a function of the independent variable.

2. Simple Linear Regression is foundation in machine learning,  serving as a basis for more complex regression models and predictive analytics.

3. Simple Linear Regression is often the first model used to solve regression problems because it is simple, easy to interpret, and provides insight into the dependency between variables.

4.  Simple Linear Regression help to understand how one factor affects another and enables prediction.

EXample :   

- predicting a student's score based on the number of hours studied.
- predict a person’s salary based on years of experience by finding the best-fit line that minimizes prediction errors.



## Question 2: What are the key assumptions of Simple Linear Regression?

Answer :  

1. **Linearity :**  The relationship between the independent variable and dependent variable is linear, meaning the change in the dependent variable is proportional to the change in the independent variable.

2. **Homoscedasticity:** The residuals have constant variance at all levels of the independent variable, meaning the error spread is uniform.

3. **Independence of errors:** The residuals (errors) are independent of each other; there is no correlation between consecutive errors.

4. **No multicollinearity [in multiple regression] :**  Though not strictly for SLR, it assumes that explanatory variables are not highly correlated in general regression contexts.

5. **Normality of residuals:** The residuals are normally distributed, which is important for inference and hypothesis testing.



##  Question 3: Write the mathematical equation for a simple linear regression model and explain each term.

Answer :   

Mathematical equation for a simple linear regression :    

y = mx+b+error

y : dependent variable(target/respone) we want to predict or explain

x : independent variable(coefficient/predictor) used to explain or predict y

m : slope

    - slope tells us, how much the dependent variable y changes, on average, for a one-unit increase in the independent variable x.
    - If the slope is +ve, it means that as x increases by 1 unit, y increases on average by the value of the slope.
    - If the slope is -ve, it means that as x increases by 1 unit, y decreases on average by the value of the slope.
    - A slope of zero indicates no linear relationship between x and y, meaning y doesn't change with x.
    - example, we want to  predict a student's score  based on the number of hours studied.

                  y = 40 + 4x  {after fitting regression model,equation}

                  here slope is 4 , This means for every additional hour studied {increase of 1 in x}, the students score y increases by 5 points.so if a student studies 1 hour more, their score is predicted to increase by 5.

                  slope quantifies the strength and direction of the relationship between hours studied and scores—it shows how much the outcome changes with a one-unit change in the predicto

b: intercept

    - it is where the regression line crosses the Y-axis. It helps anchor the regression line on the graph and is essential for making predictions.
    - The intercept in regression is the value of the dependent variable when the independent variable is zero.
    - It represents the starting point or baseline value of the predicted outcome before considering the effect of the predictor
    - example , predicting exam scores based on hours studied, the intercept is the expected score when hours studied is zero.

error :  error term, which accounts for the variation in y not explained by the linear relationship with x.

- residual = y_actual - y_predicted

- Ordinary Least Squares method :                                       

              min { summation [ y_actual - y_pred]^2}

              min { summation [ y_actual - (mx+b)]^2}

sum of squared residual between actual and predicted data points.the parameter m and b are chosen to make this sum as small as possible, which gives the best-fitting line through the data points.

- Cost Function :  mean squared error

             1/n { min { summation [ y_actual - (mx+b)]^2} }

finding the values of m and b that produce the lowest average squared error, providing the best linear fit to the data.

## Question 5: What is the method of least squares in linear regression?

Answer :  

- The method of least squares in linear regression is a way to find the line that best fits the data points by minimizing the sum of the squared differences between the observed values and the values predicted by the line. In simple terms, it finds the line where the total of all squared vertical distances (errors) from the data points to the line is as small as possible.

1. Suppose we have data points with inputs x and ouput y.
2. we want to predict y based on x using linear regression :   
                y=mx+b
3. The least squares method calculates m and b to minimize the sum of squared differences:
                Summation { y_actual - y_pred}^2

4. Minimizing this sum means the line is as close to the points as possible on average.
5. This method gives us the best linear prediction line for your data.
6. Minimizing these squared differences ensures the best linear approximation to the data, reducing the effect of large deviations by squaring them.
7. The slope and intercept are found using formulas based on summations of x_i,y-i,x_i*y_i,x_i^2, which result from setting the derivatives of the sum of squared residuals to zero to find the minimum.

- The least squares method is foundational in regression analysis, widely used for predicting outcomes, estimating trends, and modeling relationships between variables.
- In statistical terms, under assumptions such as independent, normally distributed errors with equal variance, least squares estimates provide the best linear unbiased estimators (BLUE) per the Gauss-Markov theorem.

## Question 6: What is Logistic Regression? How does it differ from Linear Regression?

Answer :  

* Logistic regression is a powerful statistical and machine learning technique that models the probability of a binary outcome based on one or more independent variables. It transforms a linear combination of inputs using a logistic (sigmoid) function, which maps any real-valued number into a probability between 0 and 1. This makes it ideal for classification tasks where the goal is to predict the likelihood of an event occurring, such as whether an email is spam or a patient has a disease. Beyond simple prediction, logistic regression provides interpretable results about the influence of each predictor on the outcome, making it a preferred method in domains like healthcare, finance, and social sciences for both prediction and insight extraction. Its mathematical foundation using maximum likelihood estimation ensures robust parameter fitting even with complex relationships and mixed data types.

**Difference**

1. Purpose :
  - Linear Regression predicts continuous numerical values (e.g., price, temperature).
  - Logistic Regression predicts probabilities of categorical outcomes, especially binary classification (e.g., spam or not spam).

2. Output :   
    - Linear Regression outputs continuous values on a straight line.
    - Logistic Regression outputs probabilities between 0 and 1 using an S-shaped sigmoid curve.

3. Equation :
   -   linear regression :  y = mx+b  ==> linear equation
   -   logistic regression : y = 1 / { 1 + e^(-mx+b)}     ==> which maps any real number to a probability.

4. loss function :          
    - Linear Regression minimizes the sum of squared errors (least squares).
    - Logistic Regression maximizes likelihood using log loss (cross-entropy loss).

5. Assumptions:
    - Linear Regression assumes a linear relationship between input and output and normally distributed residuals.
    - Logistic Regression assumes the log-odds (logit) of the outcome is linearly related to the inputs and models categorical outcomes.

6. Example :       
    - Linear Regression: price predictions, sales forecasting.
    - Logistic Regression: disease diagnosis (yes/no), email spam detection.

## Question 7: Name and briefly describe three common evaluation metrics for regression models.

Answer :   

**Mean Squared Error[MSE]:**

- MSE calculates the average of the squared differences between predicted and actual values.MSE penalizes larger errors more significantly, making it useful for identifying models that fit the majority of data points closely but may have a few substantial mistakes

MSE = 1 / n summation { (y_act - Y_pred)^2}

Advantage :  
            - Act as Cost Function
            - Ensures that positive and negative errors don't cancel out each other.
            - Its differentiable used with gradient descent.

Disadvantages:
            - Can be overly influenced by outliers due to squaring.
            - Less intuitive to interpret since units are squared.

- lower mse means better model

**Mean Absolute Error[MAE]:**

- MAE measures the average magnitude of the errors between predicted and actual values, without considering their direction. It gives an easily interpretable value in the same units as the target variable, indicating on average how much predictions deviate from true values.

MAE = 1 / n | summation { (y_act - Y_pred)} |

Advantage :                         
            - Less sensitive to outliers, so it gives a robust measure of average error magnitude.
            - It will be same unit ,its more interpretable

Disadvantage :            
             - Not differentiable at zero, which can complicate some optimization algorithms
             - Convergence usually takes time or optimization is complex so model training is time consuming.
             - Treats all errors equally regardless of their size, which can be a limitation when large errors need more attention.

**Root Mean Squared Error[RMSE]:**

1. measures the average distance between the predicted values and the actual values.
2. RMSE is the square root of the average of the squared differences (errors) between predicted and actual values.

      RMSE = sqrt [1 / n summation { (y_act - Y_pred)^2}]

- RMSE has the same units as the target variable, making it easy to interpret.
- RMSE is useful because it penalizes larger errors more (due to squaring) and gives a clear measure of model prediction accuracy in the same scale as the data.

**R-squared (R²) Score:**

- The R-squared score, also called the coefficient of determination, quantifies how much of the variance in the target variable is explained by the predictors in the model. An R² value close to 1 indicates a good fit, while a value close to 0 suggests that the model explains little of the observed variability

- Suppose we have data points scattered on a graph:   

          - The R-squared value shows how close the data points are to the line drawn by your model.

          - If R-squared is 0, the model explains none of the variation (the line doesnt fit the data).

          - If R-squared is 1, the model explains all the variation perfectly (all points lie exactly on the line).

          - For example, an R-squared of 0.75 means the model explains 75% of the variation in data, while 25% is unexplained.

- Formula :     

            R_Square = 1 -  { SSR / TSS} = 1 - unexplained_variation / Total_variation

or          R_square = Explained Variation / Total variation


1.Sum of Squared Residuals (unexplained variation) :

          SSR  = summation { y_actual - y_pred }^2

- It measures the sum of the squared differences between the observed values and the values predicted by the model (errors or residuals).
- we are not able capture difference between pred and obs or this amount of variation is not explained by predicted line.
- Smaller values indicate a better model fit.

2.Total Sum of Squares (total variation in data) :

It measures the total variation or spread in the observed data. Specifically, it is the sum of the squared differences between each observed value and the mean of the observed values.

            TSS = SSR + SSE

- Sum of Squares due to Regression (SSR) - Explained Variation

This measures the variation explained by the regression model—how much the predicted values differ from the mean of the observed values. It quantifies how well the model captures the data's variation.

                      SSR = summation { y_pred - y_avg }^2

y_avg =  mean of the observed values

- Sum of Squares for Error (SSE) or Residual Sum of Squares - Unexplained Variation

This measures the variation the model fails to explain—how much the actual values differ from the predicted values (errors or residuals).

            SSE/SSR = summation { y_act - y_pred } ^ 2

SSR hows how much of the total variation is explained by the model.

SSE shows how much variation remains unexplained (error).

TSS is the total variation in the data around the mean.

Advantage :     

1. rovides a clear metric to compare model fits.
2. Useful for assessing the goodness of fit in regression.
3. Invariant under linear transformations of the predictors, meaning it works regardless of unit scales.

Disadvantage :             

1. Not suitable for nonlinear models without transformation.
2. Sensitive to adding more variables, which can artificially increase R_square
3. Can be misleadingly high for models that overfit data.
4. high R-squared doesn't always mean a perfect or useful model—it just quantifies explained variance, not causality or predictive accuracy.

**Adjusted R-Squared:**

modified version of R-Squared that adjusts for the number of predictor variables in a regression model.

Adjusted R-Squared = 1 - { [1-R_sqaure] * (n-1)/(n-k-1) }

n = number of observations
k = number of predictor variables

1. It tells you how well the model explains the data, taking into account the number of predictors used.
2. Unlike R-Squared, Adjusted R-Squared only increases if the new predictor improves the model, and it decreases if the predictor does not add value.
3. It prevents overfitting by penalizing unnecessary variables.
4. Values range from 0 to 1, with higher values indicating a better model fit after adjustment for predictors.

- Adjusted R-Squared is useful when comparing models with different numbers of features to find the best model that balances fit and complexity.

## Question 8: What is the purpose of the R-squared metric in regression analysis?

Answer :  

R-squared is a statistical measure that indicates how well the independent variables in a regression model explain the variance of the dependent variable. It quantifies the proportion of the total variation in the outcome that is captured by the model, with values ranging from 0 to 1, where a higher value indicates a better fit. This metric is used to evaluate the goodness-of-fit of the model and to compare different models. However, R-squared alone does not guarantee that the model is appropriate, and adjusted R-squared is often preferred to account for the number of predictors and prevent overfitting.

- In practice, R-squared helps you assess how well the models predictions match the actual data, allowing you to compare competing models and choose the one that best fits your data.


## Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

Answer :   

```python
import pandas as pd

data=pd.DataFrame({
    'Height': [60, 62, 65, 68, 70, 72, 75],
    'Weight': [115, 120, 130, 140, 150, 160, 180]
})
```

```python
data.head()
```

```python
X = data['Height'].values.reshape(-1, 1)
y = data['Weight'].values
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
slope=model.coef_[0]
```

```python
print(f"Slope [coefficient] : {slope}")
```

```python
intercept = model.intercept_
```

```python
print(f"Intercept : {intercept}")
```

```python
# prediction
height_to_predict = 67
predicted_weight = model.predict([[height_to_predict]])
```

```python
print(f"Predicted weight for height 69 inches: {predicted_weight[0]} Kg")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a scatter plot of the actual data points
plt.scatter(X, y, color='blue', label='Actual Data')

# plot regression line using slope and intercept
# generate x value for the regression line within range of hegiht
x_line = np.array([X.min(), X.max()]).reshape(-1, 1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, color='red', label='Regression Line')


plt.scatter(height_to_predict, predicted_weight[0], color='green', marker='X', s=100, label=f'Predicted: {predicted_weight[0]:.2f} Kg')

plt.xlabel('Height (inches)')
plt.ylabel('Weight (Kg)')
plt.title('Height vs Weight with Simple Linear Regression Line and Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
