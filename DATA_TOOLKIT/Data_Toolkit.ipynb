### 2. How does broadcasting work in NumPy?

Answer:

> Broadcasting in NumPy is a set of advanced rules that enable arithmetic operations between arrays of different shapes without the need for explicit resizing.When two arrays have compatible shapes, NumPy intelligently expands the smaller array in memory, treating it as if it had the shape of the larger one—this allows for efficient, element-wise operations.

>example, adding a scalar to an array or combining a 1D and a 2D array is seamless due to broadcasting. The rules check each dimension from right to left: dimensions are compatible if they are equal or one is 1. This mechanism makes code concise, fast, and highly efficient, leveraging optimized C routines without unnecessary data copying. Broadcasting is a cornerstone for vectorized computations and is critical in large-scale data analysis and machine learning tasks, where speed and memory efficiency matter most.

> Rule:

  1. If two arrays have different numbers of dimensions, prepend (add) ones to the smaller array's shape on the left until both have the same number of dimensions.

  2. NumPy compares the arrays' shapes from right to left dimension by dimension.

  3. Two dimensions are compatible if:

              A] They are equal, or

              B] One of them is 1

  4. If dimensions are compatible, the array with dimension 1 is stretched (broadcast) to match the other size.

  5. If any pair of dimensions are not compatible (not equal and neither is 1), NumPy raises an error.

  6. If shapes are (4, 3) and (3,), NumPy treats the (3,) array like (1, 3), then broadcasts it to (4, 3) for operations

```python
#example_1
import numpy as np

a=np.arange(1,4).reshape(3,1)
b=np.arange(1,10).reshape(3,3)
print(a)
print("----")
print(b)

result=a+b
print("----")
print(result)
```

```python
# exampe_2 : value error
a = np.array([[1, 2], [3, 4]])
b=  b = np.array([10, 20, 30])
try:
  result = a + b                       # This will cause a ValueError
  print(result)
except ValueError as e:
  print(e)

# shape of a is (2,2) and b is (3,)
#broadcasting rule =  2 vs 3 → not equal and neither is 1 (incompatible).
#Because these dimensions dont meet broadcasting rules, NumPy cannot stretch b to match
```

### 6. What are the differences between NumPy arrays and Python lists?

Answer:

1. Data Type:

    - NumPy arrays are homogeneous, meaning all elements have the same data type (e.g., all integers or floats) for better performance.

    - Python lists are heterogeneous and can store mixed data types (ints, strings, objects).

2. Performance and Speed:

    - NumPy arrays are implemented in C and allow vectorized operations, making them much faster for numerical computations than Python lists.

    - Python lists are slower because operations happen element-by-element in pure Python.

3. Memory Efficiency:

    - NumPy arrays store data in contiguous blocks of memory leading to efficient use of space and faster access.

    - Python lists store references to objects and are less memory efficient.

4. Functionality:

    - NumPy provides a wide range of mathematical and statistical functions optimized for arrays.

    - Lists are basic containers without built-in mathematical operations.

5. Mutability and Size:

    - NumPy arrays have a fixed size; resizing requires creating a new array.

    - Python lists are dynamic and can grow or shrink easily.

### 8.  What does the term “vectorized operation” mean in NumPy?

Answer:-

> A vectorized operation in NumPy means performing mathematical or logical operations on entire arrays at once, without using explicit loops. Instead of processing elements one by one, NumPy applies the operation simultaneously on whole arrays using optimized, low-level code.

> Advantage

    - Faster performance: Vectorized operations are much faster than writing loops in Python, because NumPy uses efficient C-based implementations behind the scenes.

    - Cleaner code: It lets you write simpler and shorter code by eliminating explicit loops.

    - Handles large data: Vectorization scales well for big datasets, making NumPy ideal for numerical computing and data science

### 21.  Why is NumPy's array slicing faster than Pythons list slicing?

Answer:-

- Contiguous Memory Storage: NumPy arrays store elements in a continuous block of memory, making access and slicing operations faster. Python lists store references to objects scattered across memory, causing slower access.

- Fixed Data Type: NumPy arrays have a fixed data type for all elements, so slicing involves simple pointer arithmetic without type checks. Python lists can hold mixed types, requiring extra overhead to handle each element.

- Low-Level Optimizations: NumPy slicing is implemented in fast compiled C code, avoiding Python's interpreter overhead for loops or indexing.

- Less Overhead: NumPy slices create views on the same data without copying, while Python slicing often creates new lists, which takes more time and memory.
