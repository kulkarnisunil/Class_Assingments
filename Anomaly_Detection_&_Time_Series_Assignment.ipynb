## Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

Answer:

Anomaly Detection is the process of identifying data points, events, or patterns that deviate significantly from the expected or normal behavior in a dataset. These unusual observations are called anomalies or outliers and often indicate critical events such as fraud, system faults, cyberattacks, or medical abnormalities.

**Point Anomaly**

A point anomaly is a single data instance that is significantly different from the rest of the data its also called global or individual outliers.

Example :

* In a bank transaction dataset, most transactions are below ₹50,000, but one transaction of ₹10,00,000 appears — this transaction is a point anomaly.

* In temperature monitoring, a single reading of 90°C when the normal range is 20-30°C.

**Contextual Anomalies**

Contextual anomalies appear normal in one context but abnormal in another,such as time, location, or season.

Example :             

* A temperature of 30°C is normal in summer but anomalous in winter.

* High credit card usage at midnight may be normal during holidays but suspicious on regular days.

**Collective Anomalies**

A collective anomaly happens when:

* Each data point looks normal by itself
* But when you look at them together (as a group or sequence), the behavior is abnormal

Example :                                

* Online Shopping / Website Visits :

One user refreshes a page = Normal

Refreshes again = Normal

But:

Page refreshed thousands of times per minute

==> Looks like a bot attack
==> Collective anomaly

* Network Security :

1-2 login attempts per minute is normal

repeated hundreds of times

Each login attempt is normal,But the repeated pattern signals a brute-force attack

## Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

Answer:

**Isolation Forest**

Isolation Forest randomly splits data using trees(Randomly selects features and split values), anomalies are isolated in fewer splits, resulting in shorter path lengths compared to normal points.

Suitable for Large datasets,High-dimensional data,Point anomalies,Fraud detection, intrusion detection.

* Advantage:   

1. Very fast and scalable

2. Isolation Forest is fast and scalable because it isolates points using random splits instead of distance calculations, making it efficient for large, high-dimensional datasets.

* Disadvantage :    

1. Not ideal for sequential or collective anomalies: Isolation Forest treats each data point independently,it does not care about order or patterns over time.

2. Its anomaly decisions are based on random tree structures, making them difficult to explain in simple rules.

3. Isolation Forest is less effective for local outliers because it does not consider neighborhood density and struggles when data has clusters of varying densities


**DBSCAN**

DBSCAN finds groups of closely packed points as clusters and marks lonely points in empty areas as outliers using eps and minPts parameters.

Suitable for Spatial or geographic data,Data with arbitrary-shaped clusters,Noise detection

* Advantage:  

1. No need to specify number of clusters

2. Detects clusters of any shape

3. Naturally finds outliers

* Disadvantage :                  

1. Sensitive to eps and min_samples.

how close points must be to be considered neighbors,eps is too small ==> many points marked as outliers and eps is too large everything becomes one big cluster.

2. Struggles with varying densities.

DBSCAN uses one eps distance for all clusters.

Works fine if all clusters have similar density

Struggles if clusters have different densities

3. Poor scalability in high dimensions because Uses distance calculations and in high dimesion Distances become less meaningful,Computation becomes slow,Performance drops badly

**Local Outlier Factor**

LOF measures how dense a point is compared to its neighbors; if it's much less dense locally, it's flagged as an outlier.

Suitable for Data with local patterns,Anomalies inside clusters,Customer behavior analysis.

- Advantage

LOF detects anomalies relative to local neighborhoods, so it works well even when clusters have different densities.

- Disadvantage

1. Computationally expensive because calculating distances between each point and its neighbors For large datasets, this takes a lot of time and memory.

2. Sensitive to number of neighbors.

selecting proper k value is difficult. k=number of neighbors.


if k is too small ==> might flag normal points as anomalies

If k is too large ==> might miss anomalies

3. Because of distance calculations and neighbor comparisons, LOF does not scale well For millions of points, it becomes slow and memory-heavy for large dataset.

## Question 3: What are the key components of a Time Series? Explain each with one example.

Answer:

A time series has four components: Trend (overall direction), Seasonality (fixed repeating patterns), Cyclical (long-term fluctuations), and Irregular (random noise).

**1.Trend**

- The overall direction of the data over a long period
- Can be upward, downward, or flat
- Ignores short-term fluctuations and random noise
- Visual clue:A smooth rising or falling curve in the time series chart
- Example :             

Upward trend:

        1. Monthly sales of a store increasing from ₹50,000 → ₹80,000 over 5 years
        2. City population increases from 1 million → 1.5 million in a decade.
        3. Number of Electric vehicle registrations rises from 5,000 → 100,000 in 5 years.

Downward trend:

        1. Annual number of landline subscribers decreasing over 10 years.
        2. Number of landline subscribers decreases from 500,000 → 100,000 in 8 years.
        3. Average weekly Movie theater audience attendance decreases from 5,000 → 1,500 over 7 years due to streaming.

**2.Seasonality**

- Some events repeat regularly at fixed intervals
- Could be daily, weekly, monthly, or yearly
- Caused by seasonal factors, habits, or predictable cycles
- Visual clue:Regular peaks and troughs
- Example :-                                         

1. Daily seasonality:

Electricity usage peaks in the evening when people return home.

2. Weekly seasonality:

Restaurant sales are higher on weekends than weekdays.

3. Monthly seasonality:

Ice cream sales peak every summer.

4. Yearly seasonality:

Retail sales spike in December due to Christmas and holiday shopping.

**Cyclical Component**

- Fluctuations that happen over long periods (years).
- Unlike seasonality, the length of the cycle is not fixed(depends on economy or market like other factor).
- Often influenced by economic, business, or market conditions.
- Visual clue: Waves or swings that do not repeat exactly at regular intervals.
- Example :-        

1. Economic growth and recession:

GDP rises for several years (expansion), then falls for a few years (recession).

2. Stock market cycles:

Bull market (rising) for several years, then bear market (falling).

3. Real estate prices:

Property prices increase for a decade, then decrease for a few years.

4. Automobile industry sales:

Car sales rise during economic boom, fall during slowdown.

**Irregular / Random Component**

- These are unpredictable, one-time fluctuations in the data
- Caused by unexpected events or random noise
- Cannot be explained by trend, seasonality, or cyclical patterns
- Visual clue:Spikes or dips that don't follow a pattern
- Example :  

1. Natural disasters:

Sudden drop in airline passengers due to a hurricane or earthquake.

2. Political events:

Stock market drops unexpectedly because of an election result.

3. Equipment failure / accidents:

Factory production drops for a day due to machine breakdown.

4. Random noise:

Small random variations in daily temperature readings.

## Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

Answer:

A stationary time series is one whose statistical properties do not change over time, such as:Mean (average value),Variance (spread of data),Autocorrelation (relationship between past and future values)

Example :    

- Daily temperature in a controlled lab (mean ~25°C, small variance)
- Stock prices generally are not stationary (they trend upwards or downwards)

- Stationarity is important because stationary series allow models to learn patterns from the past and make accurate, reliable forecasts.

**Testing for Stationarity**

1. Visual Inspection : Visual inspection involves plotting the time series and checking if it has a roughly constant mean, constant variance, and no obvious trend or seasonality.

2. Summary Statistics : Summary statistics involve splitting the series into two halves and comparing the mean and variance to check if they remain roughly constant over time.

3. Statistical Tests :

      - Augmented Dickey-Fuller (ADF) Test :

        Null hypothesis: Series is non-stationary

        p-value < 0.05 → Reject null → Series is stationary


     - KPSS Test (Kwiatkowski-Phillips-Schmidt-Shin)
     
        Null hypothesis: Series is stationary

        p-value < 0.05 → Reject null → Series is non-stationary

**Transforming to Stationarity**

1. Differencing :  y_dast_t = y_t - y_t-1

- Differencing subtracts the previous observation from the current one to remove trends and make a time series stationary.
- This removes long-term trends in the data and helps stabilize the mean.

2. Log Transformation : y_dash_t = log(y_t)

- Log transformation applies the logarithm to each data point to stabilize variance and reduce the effect of large values in a time series.
- Helps stabilize variance, especially when the series has values that grow exponentially or have large fluctuations.
- Log-transformed series reduces the spread, making it easier to model

3. Detrending : (Yt - Trend)       

- Remove the long-term trend from a series so that the remaining data fluctuates around a constant mean.
- Can be done using:

      - Regression: Fit a trend line (linear or polynomial) and subtract it from the original series

      - Moving Average: Smooth the series to find the trend, then subtract it

- Subtracting the trend gives stationary fluctuations because allow models to learn genuine patterns instead of overall growth.

example : If sales keep increasing every year, the model may think future sales will always rise.Removing the trend lets the model learn true behavior, not just growth

4. Seasonal differencing :   

-Seasonal differencing removes repeating seasonal patterns by subtracting the value from the same period in the previous cycle.

- Current value minus value from the previous season.
- Models like ARIMA don't work well with this repetition.Seasonal differencing removes this repetition and keeps only the real change over time.

## Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

Answer:

AR and MA handle stationary data, ARIMA handles trend, SARIMA handles seasonality, and SARIMAX also considers external influencing factors.

**AR(AutoRegressive)**

AR model predicts the current value using its own past values and is suitable for stationary time series with strong autocorrelation.

Auto → uses its own past values

Regressive → predicts the current value using a regression equation

y_t= c + ϕ_1 * y_t-1 + ϕ_2 * y_t-2 + ϵt

y_t = Value of the time series at current time t

c=constant/intercept

-Represents the mean level of the series and Keeps the model from always predicting zero.

ϕ_1,ϕ_2 = Autoregressive coefficient

- Measure how much influence past values have
- Larger value → stronger influence
- Can be positive or negative

y_t-1,y_t-2 = Lagged values (previous observations)

-like y_t-1 : yesterday ,y_t-2=day before yesterday

p = Order of the model

p = 1 → simple, short memory

p large → complex, long memory

How to choose p:PACF (Partial Autocorrelation Function),Information criteria (AIC, BIC)


ϵt=(White Noise Error)

- Random shock or noise
- Mean = 0, constant variance
- Represents unpredictable events

* 3. Lag Concept (Why Past Values Matter)

Lag = previous time step

AR uses lagged values as predictors

Example:

AR(1): uses only yesterday's value

AR(2): uses yesterday and the day before

More lags → capture longer memory of the series

* Assumptions

1. Stationary time series
- Mean and variance do not change over time
2. Linear relationship
3. Errors are independent and identically distributed


* Advantages :         

1. Simple and interpretable
2. Good for short-term forecasting
3. Works well for stable time series

* Disadvantages :      

1. Cannot handle trend
2. Cannot handle seasonality
3. Poor performance on non-stationary data
4. Sensitive to wrong lag selection

* Aplication :     
1. Stock prices (short-term)
2. Weather temperature
3. Sensor data
4. Economic indicators
5. Works only with stationary data (constant mean & variance).
6. Captures trend patterns, but cannot model shocks/noise well.

Example: Stock price movement based on previous days

**MA (Moving Average) Model:MA(q)**

MA (Moving Average) models a time series by using past errors (shocks) or residuals to predict the current value.

- Learn from past prediction mistakes to improve today's forecast
- It focuses on how past unexpected changes affect the present

        MA handles short-term randomness in the data
        
        If something unexpected happened yesterday, it influences today

        Example:Daily traffic spike due to an accident yesterday - affects prediction for today,MA captures this effect.

-Not to be confused with the simple moving average (smoothing technique)

Simple moving average (SMA) = average of past k values - used to smooth data

MA in ARIMA/SARIMA = uses past errors → used to model time-series and forecast

            EXample :

            SMA: Average sales last 3 days = smoothing

            MA: Adjust todays forecast based on past 3 day's forecast errors = modeling
  
* Why MA is useful :      

- MA captures short-term random shocks that are not captured by AR.
- MA reduces bias in forecasts by learning from past prediction errors.

* formula = yt=c+ϵt+θ1ϵt-1

y_t = current value,ϵt=current random error,θ1 = weights for past errors

* Advantages:

1. Handles sudden changes in the data that AR cannot(Captures short-term shocks).

      Example: Sales spike due to a one-day promotion

2. Uses past forecast errors to improve accuracy(Reduces forecast bias)

        Example : Corrects predictions if previous forecasts underestimated/overestimated

3. Easy to combine with AR and differencing in ARIMA models

* Disadvantages :  

1. Only considers a finite number of past errors (q terms).Cannot capture long-term trends or seasonality.
2. MA models errors, not the actual values,Needs AR or differencing to model trend.
3. Choosing the right q (number of past errors) is critical.

        Wrong q → overfitting or underfitting

* Example : forecasting daily temperature:

Yesterday's forecast was off by +2°C - today's forecast is adjusted to correct for that error

*Application:

1.Captures short-term effects of sudden changes.
2. Works only for stationary data.


**ARIMA (Autoregressive Integrated Moving Average)**

Current value = Mean + weighted sum of past forecast errors(Helps correct for past mistakes in predictions)

ARIMA predicts future values using past data, past errors, and trend removal, making it suitable for non-seasonal time-series forecasting.

ARIMA = learn from past values + fix past mistakes + remove trend

Autoregressive (AR) = using past observations
- AR part looks at previous values to predict the current value.
- Example: Today's sales depend on sales from yesterday and last week

Moving Average (MA) — using past errors
- MA part looks at past mistakes of the model and adjusts predictions
- Example: If last month's prediction underestimated sales, MA corrects for it.

Integrated (I) — differencing to remove trends
- Differencing removes long-term upward or downward trends
- Example: If sales are steadily increasing, we subtract previous month's value to focus on changes, not the trend


*  When to use :       
1.
- ARIMA works best when the data has a long-term upward or downward movement that needs to be modeled.If the data is mostly flat with no trend, ARIMA is not necessary.
- Example : Monthly sales increasing every year

2.
- ARIMA cannot handle repeating patterns (seasonality) well.If the data has strong seasonal spikes (e.g., Christmas sales every December), SARIMA is better.
- ARIMA is suitable for non-seasonal time-series data.
- Example :             

      Daily website traffic that jumps every weekend = ARIMA may fail

      Monthly revenue with no repeating monthly spikes = ARIMA works

3.
- ARIMA assumes the future depends on the past.If historical trends and fluctuations are useful predictors, ARIMA can model them.
- Example : Last year's sales data helps predict next month's sales

* Advantages :            

- ARIMA is simple to understand and its components are easy to explain to business teams.
- ARIMA performs well for non-seasonal data with trends and short-term variations.
- ARIMA is often used as a starting point before trying more complex models like SARIMA, SARIMAX, or LSTM.Provides a benchmark to compare model performance.

      Example:  Forecasting sales or electricity consumption: start with ARIMA, then try SARIMA if seasonality exists

* Disadvantages :      

- It cannot capture repeating patterns (seasonality) in the data.If your data has seasonal spikes, you need SARIMA instead.

        example : Electricity usage increases every summer : ARIMA cannot model this seasonal effect
- ARIMA requires careful selection of p, d, and q to achieve accurate forecasting.
- ARIMA assumes the current value changes linearly with past values and past errors.Cannot handle non-linear relationships in the data.

        Example :

              Sales increase gradually with advertising ==> linear ==> ARIMA works

              Sales jump drastically only after a threshold ==> nonlinear ==> ARIMA may fail
* Application:

1. Handles trends via differencing.
2. Widely used for short-term forecasting of time series without strong seasonality.

**SARIMA (Seasonal ARIMA)**

SARIMA (Seasonal AutoRegressive Integrated Moving Average) is a time-series forecasting model that extends ARIMA by capturing trend, seasonality, and random noise in data that shows repeating patterns over fixed time intervals.

- It is a model used to predict future values using past data collected over time.like daily sales, monthly electricity usage
- It can model long-term increase or decrease in data(Sales growing every year).
- It can model repeating patterns at fixed intervals(Higher sales every December,Higher electricity usage every summer)
- Even after trend and seasonality, data still has small, unpredictable ups and downs.These are called random noise or random shocks.SARIMA models these using the MA (Moving Average) part, it learns from past shocks to reduce their effect on future predictions.
- Seasonality refers to patterns that repeat at fixed and known intervals, such as weekly or yearly cycles.

* Model Notation:  

SARIMA(p,d,q)(P,D,Q,s):  

p,d,q = non-seasonal AR, differencing, MA
P,D,Q = seasonal AR, differencing, MA
s = length of the seasonal cycle

1. Why SARIMA is Used:

- ARIMA handles trend but cannot handle seasonality
- SARIMA explicitly models repeating seasonal patterns. like Sales this December depend on sales last December
- Example : Monthly sales data,Electricity consumption,Airline passenger traffic

2. Assumptions :  

- Data becomes stationary after differencing
- Residuals are white noise
- Relationship is linear

3. Advantage :      

- SARIMA can capture both long-term trends and repeating seasonal patterns in time-series data.
- SARIMA is a classical statistical model whose components and coefficients are easy to understand and explain, like Sales increase by 2 units for each unit increase in last month's sales .
- SARIMA works well on many real-world datasets Even if more complex models exist (like LSTM), SARIMA is a reliable starting point for comparison so SARIMA is often used as a baseline model for time-series forecasting due to its proven accuracy and reliability.exmple Forecasting electricity consumption,Forecasting airline passengers

4. Disadvantages :      

- SARIMA only uses past values, past errors, and seasonal patterns to make predictions.It cannot directly use external factors like marketing spend, temperature, or holidays.
- SARIMA is computationally more intensive than ARIMA due to additional seasonal components.
- Requires careful parameter tuning

* Application:

1. Ideal for seasonal time series (electricity usage, sales, weather patterns).
2. Captures both trend and repeating seasonal patterns.

**SARIMAX**

SARIMAX (Seasonal AutoRegressive Integrated Moving Average with eXogenous variables) is a time-series forecasting model that extends ARIMA by handling trend, seasonality, and external influencing variables together in a single framework.

SARIMAX uses past values(AR-learn from past values) and past errors(learn from past error) and applies differencing(remove trend to make data stationary) to handle non-stationary data.like model uses history + mistakes + trend removal to forecast accurately.

SARIMAX can model recurring seasonal patterns such as monthly, weekly, or yearly cycles using seasonal AR, MA, and differencing terms.

SARIMAX improves forecasting by incorporating external variables that affect the time series.These external variables are called exogenous variables (X) and are added alongside time-based patterns.

SARIMAX is considered the most powerful classical model because it jointly models trend, seasonality, noise, and external influence.

* Notation :   

SARIMAX(p,d,q)(P,D,Q,s)+X

p,d,q=Non-seasonal AR, differencing, MA

P,D,Q = Seasonal AR, differencing, MA

s = Length of season

X = Exogenous (independent) variables

* Role of Exogenous Variables :  

- Exogenous variables represent external influences that impact the target time series.like Exogenous variables represent external influences that impact the target time series.

- Exogenous variables should not be dependent on the target variable itself.

- Future values of exogenous variables must be known or forecasted to make predictions.like Planned marketing budget,Scheduled price discounts,is not good for Future customer sentiment (unknown)

1. Assumptions

- Residuals should be white noise :   Residuals are the part of the data not explained by the model. In a good SARIMAX model, residuals behave like white noise—random, patternless, with zero mean and constant variance. If residuals still show trend or seasonality, it means the model is missing important information.

- SARIMAX assumes a linear relationship between the target and its predictors.

- Exogenous variables should be independent of model errors; otherwise, the model confuses reactionary effects with true causal impact, leading to biased forecasts.

2. Advantages:       

- SARIMAX provides clear, interpretable coefficients that explain the impact of each factor.

- SARIMAX directly models real-world drivers instead of relying only on historical patterns.

- When external variables significantly affect the target, SARIMAX outperforms ARIMA and SARIMA.


3. Disadvantgaes :              

- SARIMAX requires careful selection of parameters to avoid overfitting or underfitting.
- SARIMAX depends on future availability of exogenous variables for forecasting.
- SARIMAX is computationally more expensive due to its complexity.

* Application:

1. Forecasting where exogenous factors are important.
2. Often used in energy, retail, finance where external predictors improve accuracy.

## Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/NYC_taxi_fare_data.csv")
```

```python
data.head()
```

```python
# Abnormal taxi trips are detected using numerical features like fare amount and trip distance

#1. fare amount = It shows how much the passenger paid

#2. trip distance = It shows how far the taxi traveled

#Learn what normal trips look like, Identify trips that are rare or very different from most data points and Each trip is labeled as: Normal,Anomalous (abnormal)
# Long distance = high fare,Short distance = low fare,If this relationship is broken, it signals an anomaly.
# Example anomalies: Very high fare but short distance, Very long distance but very low fare
```

```python
features = ['fare_amount', 'trip_distance']
X = data[features]
```

```python
X.head()
```

```python
# Train Isolation Forest
iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.01,   # 2% of trips are assumed to be abnormal
    random_state=42
)
```

```python
iso_forest.fit(X)
```

```python
# Fit and predict anomalies
data['anomaly'] = iso_forest.predict(X)
```

```python
data['anomaly'].value_counts()
```

```python
# Separate normal and anomalous points
normal = data[data['anomaly'] == 1]
anomalies = data[data['anomaly'] == -1]
```

```python
data.head()
```

```python
plt.figure()
plt.scatter(normal['fare_amount'], normal['trip_distance'], label='Normal')
plt.scatter(anomalies['fare_amount'], anomalies['trip_distance'], label='Anomaly')
plt.xlabel('Fare Amount')
plt.ylabel('Trip Distance')
plt.title('Isolation Forest – NYC Taxi Fare Anomaly Detection')
plt.legend()
plt.show()
```

```python
# Normal trips show a clear pattern: as trip distance increases, fare amount also increases. Most trips form a dense cluster.

# Anomalies lie far from this pattern:High fare for short distance,Low fare for long distance,Extreme values of fare or distance outside the
# usual range
```

```python
#Isolation Forest clearly finds trips that do not follow the normal fare–distance pattern.
#These unusual trips may indicate data errors, possible fraud, or special fare cases.
```

## Question 8: Train a SARIMA model on the monthly airline passengers dataset.Forecast the next 12 months and visualize the results.

Answer:

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
data=pd.read_csv("/content/AirPassengers.csv")
```

```python
data["Month"]=pd.to_datetime(data["Month"])
```

```python
data.set_index("Month",inplace=True)
```

```python
ts=data["#Passengers"]
```

```python
# Training SARIMA model
model = SARIMAX(
    ts,
    order=(1, 1, 1),              # ARIMA(p,d,q)
    seasonal_order=(1, 1, 1, 12)  # SARIMA(P,D,Q,s)
)
```

```python
results = model.fit()
```

```python
results
```

```python
# Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()
```

```python
plt.figure()
plt.plot(ts, label="Observed Data")
plt.plot(forecast_mean, label="Forecast (Next 12 Months)")
plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    alpha=0.3
)
plt.title("SARIMA Forecast for AirPassengers")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

```python
# SARIMA learns past trend, seasonality, and patterns,It uses yearly seasonality (12 months)
# Forecast shows increasing passenger demand with seasonal ups and downs
```

```python
# Observation :
# 1. The forecast follows the historical upward trend
# 2. Seasonal peaks continue in future months
# 3. Confidence interval widens → uncertainty increases over time
```

```python
#Conclusion:
# 1. SARIMA successfully captures trend + seasonality
# 2. Suitable for monthly airline passenger forecasting
# 3. Can be used for capacity planning and demand prediction
```

SARIMA / SARIMAX)?

* How would you validate and monitor the performance over time?

* How would this solution help business decisions or operations?


Answer :

#### 1. Data Collection & Preprocessing

data source : real-time consumption data collected every 15 minutes. The dataset includes featureslike timestamp, region, weather conditions, and energy usage.

Steps:

Convert timestamps to datetime and create features like hour, day of week, seasonality indicators.

Handle missing values (interpolation for time series, forward-fill).

Encode categorical variables (region) if needed.

Scale numeric features for anomaly detection (LOF, Isolation Forest).

#### 2. Anomaly Detection in Streaming Data

Aim : Detect abnormal spikes or drops in energy usage in real-time.

Options:

1. Isolation Forest (preferred for high-dimensional streaming data):

Good for real-time, scales well, unsupervised.

Can detect sudden spikes/drops without assuming a distribution.

2. Local Outlier Factor (LOF):

Works on local density, sensitive to neighbors.

Might be slower for very large streams.

3. DBSCAN:

Density-based clustering, good for offline analysis.

Not ideal for streaming as it is batch-oriented.

* Implementation:

Use a sliding window (last N readings) for real-time detection.

Flag anomalies immediately for operational alerts.

#### 3. Short-Term Forecasting (Time Series)

Aim : Forecast energy demand for next 1-6 hours to optimize grid load.

Recommended models:

ARIMA: Good for stationary series without seasonality.

SARIMA: Better if there is daily/weekly seasonality in energy usage (common in power grids).

SARIMAX: If you want to incorporate exogenous variables like weather conditions, region, holidays.

Process:

Check stationarity (ADF test), apply differencing if needed.

Tune (p,d,q)(P,D,Q,s) for seasonality.

Forecast next time points and compare against actual usage.

#### 4. Validation & Monitoring

Validation:

Split data chronologically, e.g., last 1-2 weeks as test set.

Metrics:

Forecasting: MAE, RMSE, MAPE.

Anomaly detection: precision, recall, F1 if labeled anomalies exist.

Backtest forecasts over historical periods to ensure reliability.

Real-time monitoring:

Track rolling errors of forecasts.

Retrain model periodically if patterns change (seasonality shift, new regions).

Monitor anomaly rate: sudden spikes may indicate sensor failure vs true anomaly.


#### 5. Business & Operational Benefits

* Proactive load management: Detecting spikes allows preemptive resource allocation (e.g., turning on backup generators).

* Cost savings: Forecasting prevents overloading expensive peaking plants and reduces energy wastage.

* Reliability & customer satisfaction: Early anomaly detection prevents blackouts or grid failures.

* Operational alerts: Real-time dashboards can notify grid operators for immediate action.

* Planning & optimization: Forecasting helps in maintenance scheduling, pricing strategies, and energy trading decisions.
