# Question 1: Compare and contrast NLTK and spaCy in terms of features, ease of use,and performance.

## Answer:

Natural Language Processing (NLP) libraries are essential for analyzing and processing text data. Two of the most popular Python libraries used in NLP are NLTK (Natural Language Toolkit) and spaCy. Although both libraries are used for similar tasks, they differ in their design, usability, and efficiency.

### 1. Overview and Purpose

NLTK is one of the oldest NLP libraries and is mainly designed for educational and research purposes. It provides a wide range of tools that help users understand the basic concepts of NLP.

On the other hand, spaCy is a modern NLP library built specifically for real-world applications. It focuses on providing fast and efficient processing of large amounts of text data.

### 2. Features and Capabilities

NLTK offers a large collection of features such as tokenization, stemming, lemmatization, parsing, and text classification. It also includes many datasets and allows users to experiment with different algorithms.

spaCy provides advanced NLP features like Named Entity Recognition (NER), part-of-speech tagging, and dependency parsing. It comes with pre-trained models and supports word embeddings, making it suitable for complex NLP tasks.

### 3. Ease of Use

NLTK is flexible but requires more coding effort. Users often need to combine multiple functions to perform a complete task, which can make it slightly difficult for beginners.

spaCy is more user-friendly and provides a simple and clean API. Many NLP tasks can be completed in just a few lines of code, making it easier to use, especially for developers.

### 4. Performance

NLTK is slower because it is mainly written in Python and is not optimized for speed. It works well for small datasets and learning purposes.

spaCy is much faster because it is built using Cython and optimized for performance. It can efficiently process large datasets, making it ideal for production use.

### 5. Flexibility vs Practical Use

NLTK provides more flexibility and allows users to explore different NLP techniques in detail. This makes it a good choice for learning and experimentation.

spaCy focuses more on practical usage. It may offer fewer options than NLTK, but it provides fast and accurate results for real-world applications.

NLTK is best suited for learning, research, and experimentation, while spaCy is better for building real-world applications that require speed and efficiency. The choice between the two depends on the user's goal.

#Question 4: Describe the architecture and functioning of a Recurrent Natural Network (RNN).


## Answer:

A Recurrent Neural Network (RNN) is a type of artificial neural network designed specifically for processing sequential data such as text, speech, or time-series data. Unlike traditional neural networks, RNNs have the ability to remember previous inputs by maintaining a hidden state, making them highly suitable for tasks where context and order are important.

### 1. Architecture of RNN

The basic architecture of an RNN consists of the following components:

- **Input Layer:** Takes sequential data as input (e.g., words in a sentence).
- **Hidden Layer:** Processes the input and also receives information from the previous time step.
- **Output Layer:** Produces the final output at each time step.

The key feature of an RNN is the **recurrent connection**, where the output of the hidden layer at time step (t-1) is fed back into the network at time step (t). This creates a loop-like structure, allowing the network to retain information over time.

### 2. Working / Functioning of RNN

The functioning of an RNN can be understood step by step:

1. The network takes the first input (xв‚Ѓ) and processes it to produce a hidden state (hв‚Ѓ).
2. This hidden state (hв‚Ѓ) is stored and passed to the next step.
3. When the next input (xв‚‚) is received, the network uses both xв‚‚ and hв‚Ѓ to compute the new hidden state (hв‚‚).
4. This process continues for all time steps in the sequence.

In this way, the RNN maintains a memory of previous inputs through hidden states.

The hidden state at time step (t) is calculated as:

- \(b\) = bias  
- \(f, g\) = activation functions  

### 4. Key Characteristics

- **Memory of Previous Inputs:** Stores past information using hidden states  
- **Parameter Sharing:** Same weights are used across all time steps  
- **Sequence Handling:** Can process variable-length input sequences  

### 5. Applications of RNN

- Language modeling and text generation  
- Machine translation  
- Speech recognition  
- Sentiment analysis  
- Time-series forecasting  

### 6. Limitations

- **Vanishing and Exploding Gradient Problem** during training  
- Difficulty in learning long-term dependencies  
- Slower training compared to feedforward networks  

To overcome these issues, advanced variants like LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit) are used.


RNNs are powerful neural networks designed for sequential data processing. Their ability to maintain memory through hidden states makes them suitable for tasks where context is important. However, due to training challenges, improved architectures like LSTM and GRU are often preferred in practical applications.


# Question 5: What is the key difference between LSTM and GRU networks in NLP applications?

## Answer:

LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit) are advanced types of Recurrent Neural Networks (RNNs) designed to overcome the limitations of traditional RNNs, especially the problem of vanishing gradients and difficulty in learning long-term dependencies. While both architectures use gating mechanisms to control the flow of information, they differ in structure, complexity, and performance.

### 1. Basic Idea

Both LSTM and GRU are designed to remember important information for long periods and forget irrelevant data. They achieve this using gates that regulate the flow of information through the network.

### 2. Architecture Difference

**LSTM:**
- Contains **three gates**:
  - Input Gate (controls new information)
  - Forget Gate (decides what to remove)
  - Output Gate (controls output)
- Has a separate **cell state** and **hidden state**
- More complex structure

**GRU:**
- Contains **two gates**:
  - Update Gate (combines input and forget functions)
  - Reset Gate (controls how much past information to forget)
- Does not have a separate cell state (only hidden state)
- Simpler and more compact structure

### 3. Key Differences

- **Complexity:** LSTM is more complex, while GRU is simpler  
- **Number of Gates:** LSTM has 3 gates; GRU has 2 gates  
- **Memory Handling:** LSTM maintains separate cell state; GRU merges it with hidden state  
- **Training Speed:** GRU trains faster due to fewer parameters  
- **Performance:** LSTM may perform better on very long sequences, while GRU performs well on smaller datasets  

### 4. Working Insight

- In LSTM, information flows through a **cell state**, which acts like a long-term memory.
- In GRU, the network directly updates its hidden state without maintaining a separate memory cell.

This makes GRU computationally efficient, while LSTM provides more control over memory.

### 5. Use Cases in NLP

**LSTM is preferred when:**
- Handling long sequences (e.g., long documents, language modeling)  
- More precise control over memory is required  

**GRU is preferred when:**
- Faster training is needed  
- Dataset is smaller  
- Simpler model is sufficient  

### 6. Advantages and Limitations

**LSTM:**
- Better at capturing long-term dependencies  
- More accurate in complex tasks  
- Slower and computationally expensive  

**GRU:**
- Faster and efficient  
- Requires fewer parameters  
- May be slightly less powerful for very long sequences  

the key difference between LSTM and GRU lies in their architecture and complexity. LSTM provides a more detailed memory mechanism with three gates and a separate cell state, while GRU simplifies the design with fewer gates and faster computation. The choice between them depends on the problem size, dataset, and performance requirements in NLP applications.

# Question 8: Implement a basic LSTM model in Keras for a text classification task using the following dummy dataset. Your model should classify sentences as either positive(1) or negative (0).
# Dataset
texts = [

вЂњI love this projectвЂќ, #Positive

вЂњThis is an amazing experienceвЂќ, #Positive

вЂњI hate waiting in lineвЂќ, #Negative

вЂњThis is the worst serviceвЂќ, #Negative

вЂњAbsolutely fantastic!вЂќ #Positive

]

labels = [1, 1, 0, 0, 1]

Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on this data. You may use Keras with TensorFlow backend.


```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
texts = [
    "I love this project",                 # Positive
    "This is an amazing experience",       # Positive
    "I hate waiting in line",              # Negative
    "This is the worst service",           # Negative
    "Absolutely fantastic!"                # Positive
]
```

```python
labels = [1, 1, 0, 0, 1]
```

```python
#tokenization
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(texts)
```

```python
sequences = tokenizer.texts_to_sequences(texts)
```

```python
# Padding sequences (to make equal length)
max_length = max(len(seq) for seq in sequences)
```

```python
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')
```

```python
#Convert labels to numpy array
labels = np.array(labels)
```

```python
#Build LSTM model
model = Sequential()
```

```python
# Embedding layer (converts words to dense vectors)
vocab_size = len(tokenizer.word_index) + 1
```

```python
model.add(Embedding(input_dim=vocab_size, output_dim=16, input_length=max_length))
```

```python
# LSTM layer
model.add(LSTM(16))
```

```python
# Output layer (binary classification)
model.add(Dense(1, activation='sigmoid'))
```

```python
#Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#Train model
model.fit(padded_sequences, labels, epochs=20, verbose=1)
```

```python
#Test on new sentence
test_text = ["I really love this service"]
```

```python
test_seq = tokenizer.texts_to_sequences(test_text)
```

```python
test_pad = pad_sequences(test_seq, maxlen=max_length, padding='post')
```

```python
prediction = model.predict(test_pad)
```

```python
print("\nTest Sentence:", test_text[0])
print("Predicted Sentiment:", "Positive" if prediction[0][0] > 0.5 else "Negative")
```

- Positive sentences are labeled as 1
- Negative sentences are labeled as 0

This is a supervised learning problem.

### 2. Tokenization

The Tokenizer converts text into numerical sequences:

Example:

### 3. Padding

Since sentences have different lengths, padding is applied:
- Ensures all sequences have the same length
- Short sequences are padded with zeros

This is necessary for input into neural networks.

### 4. Model Architecture

- **Embedding Layer:** Converts word indices into dense vectors (captures meaning)
- **LSTM Layer:** Learns sequence patterns and context in text
- **Dense Layer:** Outputs final prediction (positive/negative)

### 5. Training

- The model is trained using binary cross-entropy loss
- Adam optimizer is used for efficient learning
- Accuracy is used as evaluation metric

### 6. Prediction

A new sentence is passed through the same preprocessing steps and fed into the model to predict sentiment.

### 7. Conclusion

This implementation shows how LSTM can be used for text classification by capturing sequential patterns in language. Even with a small dataset, the pipeline demonstrates key NLP steps: tokenization, padding, embedding, and sequence learning. In real-world applications, larger datasets and more complex models improve performance.
