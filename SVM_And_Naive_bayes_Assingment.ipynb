## 1. What is a Support Vector Machine (SVM)?

Answer :

SVM is a supervised learning algorithm that finds the optimal boundary (hyperplane/best possible line) to separate different classes of data points with the largest possible gap or margin between them, using support vectors (closest points).It is powerful for handling complex and small datasets, and can be used for both classification and regression tasks in machine learning.

* The SVM looks for the line (in higher dimensions, a plane) that best splits the data into groups.
* The closest data points to this line are called support vectors, and they help decide exactly where the boundary should be
* Works well for both linear and non-linear data by using a technique called the kernel trick, which lets it find boundaries even if data can't be separated simply.
* Handles small and high-dimensional datasets efficiently

formula = w^t *x + b = 0

maximize the margin (distance) between the classes, and the margin is defined using the closest points (support vectors):

margin = 2/|w|

for classification :                               

w^t * x + b > 0 for positive class

w^t * x + b < 0 for negative class

## 2. What is the difference between Hard Margin and Soft Margin SVM?

Answer :

1. Hard Margin SVM aims for a perfect separation with zero errors, suitable only for clean, perfectly separable data, but it fails{overfit} with noise or overlaps.

            formula :

            objective : min 1/2 ||w||^2

            subject to : y_i (w * x_i - b) >=1 for all x_i

            SVM tries to find a hyperplane w * x_i - b= 0 ,that separates the classes perfectly with no errors, maximizing the margin.

2. Soft Margin SVM, however, smartly allows some errors to achieve a better generalization, balancing margin size and misclassification, which makes it more robust and practical for real-world data.It uses a penalty parameter to control how much error is allowed.


            formula :                    

            objective : min 1/2 ||ww||^2 + C summation {slack_variable_i}

            subject to :  y_i (w * x_i - b) >= 1 - slack_variable_i,  slack variables allowing some errors.

            C is a penalty parameter balancing margin size and classification error.

            Soft Margin allows some misclassifications (errors) for better generalization when data is not perfectly separable.

            slack margin :
            1. measures how much a data point violates the margin or is misclassified.
            2. always greater than or equal to zero.
            3. slack_variable = 0, means data point is correctly classified and outside or on the margin boundary.
            4. slack_variable > 0, means point is either inside the margin or misclassified, allowing some flexibility in Soft Margin SVM to handle real-world noisy data.


## 3. What is the mathematical intuition behind SVM?

Answer :

SVM mathematically finds the optimal hyperplane by maximizing the margin between classes, using support vectors which are the closest points. It balances between margin width and classification errors with slack variables and a penalty parameter C. Softer margins allow better handling of noisy data. Overall, it solves a convex optimization problem efficiently to achieve powerful classification

Suppose we have two points (-4,0) and (4,4), slope = -1

Equation of hyperplane: y = w^t * x + b

Assume b = 0 , w = [−1,0] (because slope = -1)

For x = [-4 ,0], y=4>0 (below line, class +1)

For x = [4,4], y=-4 < 0 (above line, class -1)


- Points with positive y are one group, negative y another. SVM classifies based on the sign.

**Margins and Support Vectors**

Marginal planes are:

w^t * x + b = 1 and w^t * x + b = -1

These planes are parallel to the separating hyperplane and touch the nearest points (support vectors) of each class.

The distance between these planes is:

2/||w||

Maximizing this distance is the goal — it gives the maximum margin and best generalization.

**Constraints for Perfect Classification (Hard Margin)**

y_i * { w^t * x + b } >= 1

- If violated, it indicates misclassification.

- The optimal solution maximizes margin by minimizing ∥w∥ while satisfying constraints.

**Soft Margin for Noisy Data**

min ||w||^2 / 2  + C summation {slack_variable}

C is the regularization/penalty parameter balancing margin size and error tolerance.

High C means fewer errors but risks overfitting.

Low C allows more errors for better generalization (avoids overfitting).




## 4. What is the role of Lagrange Multipliers in SVM?

Answer :

Lagrange Multipliers help solve the SVM optimization problem when there are constraints (like classifying points correctly with a margin). They let us combine the goal (maximize margin) and the constraints into one function called the Lagrangian.

Instead of directly solving a complex constrained problem, we solve this combined function. The multipliers tell us how important each data point is in forming the boundary. Only points with non-zero multipliers (called support vectors) actually affect the final hyperplane.

So, Lagrange Multipliers make the math easier and help find the best dividing line by focusing on the most important points.

This approach also allows using kernel tricks for non-linear data by working only with dot products of data points.

## 7. What is a Support Vector Regressor (SVR)?

Answer :  

SVR tries to find a function (like a line or curve) that predicts target values as closely as possible while keeping the function as flat (simple) as it can. It does this by allowing some errors within a margin of tolerance (called epsilon-insensitive tube) and only penalizes errors that go beyond this margin. Support vectors are the data points that lie outside this margin and influence the regression function.

y = w^t * x + b

the minimizing :             

min 1/ 2 ||w||^2 + C summation {ξ_i + ξ_i*}


subject to :                                   

y_i - (w^t * x + b) <=  ξ + ξ_i*

(w^t * x + b)- y_i <= ξ + ξ_i*

ξ_i + ξ_i* > 0

 ξ  defines the margin of tolerance (epsilon tube) around the predicted function where no penalty is given for errors.

 ξ_i measures how far a data point is above the epsilon margin (when prediction is too low).

 ξ_i* measures how far a data point is below the epsilon margin (when prediction is too high).

are slack variables measuring the deviation outside this margin.

 ξ_i + ξ_i* indicates the total error beyond the tolerance margin for a point, and SVR tries to keep this sum as small as possible to balance accuracy and generalization.

* SVR balances model simplicity (minimizing ||w||^2 and prediction error (penalizing errors outside epsilon margin).

* It relies only on support vectors, data points outside the epsilon tube, making the model efficient.

* Kernels can be used to model nonlinear regression problems by projecting inputs into higher-dimensional spaces.

* SVR is powerful for regression where slight deviations are acceptable but outliers need penalty.

## 8. What is the Kernel Trick in SVM?

Answer :  

It transforms the original data into a higher-dimensional space where the classes can be separated by a straight line (or hyperplane). But instead of explicitly calculating this transformation (which can be very complex and slow), the kernel trick calculates the similarity between points directly in the original space.

This saves a lot of computation and allows SVM to create complex, curved decision boundaries while still using a linear classifier in that higher-dimensional space.

It works by computing dot products implicitly, enabling complex decision boundaries with less computation.

Types :             

1. Linear Kernel: Use when your data is linearly separable or has many features (high-dimensional). It's simple, fast, and effective for text or document classification—for example, spam detection.

2. Polynomial Kernel: Best for data with polynomial relationships or curves. Useful in image recognition or computer vision tasks where patterns follow polynomial trends.

3. Radial Basis Function (RBF) Kernel: Default choice for most non-linear problems. It handles complex data shapes like spirals or clusters without prior knowledge of the data's structure, such as in bioinformatics for protein classification.

4. Sigmoid Kernel: Similar to neural network activation function, good for problems where data behaves like signals or you want to mimic neural networks. Less common but useful in some specialized cases.

## 10. What is the effect of the C parameter in SVM?

Answer :

The C parameter in SVM controls the balance between having a smooth decision boundary and correctly classifying all training points.

When C is large, the SVM tries hard to classify every training point correctly, even if it means a complex boundary that fits closely to the data. This can lead to overfitting.

When C is small, the SVM allows some misclassifications but aims for a wider margin, which gives a simpler and more generalized model that may underfit the training data.

## 11. What is the role of the Gamma parameter in RBF Kernel SVM?

Answer :

formula = exp ( -γ*||x_i -x_j||^2)

γ (gamma) controls how quickly the similarity between points x_i and x_j decreases as they move further apart.

* Gamma controls how far the influence of a single training point reaches.
* Low gamma means the influence spreads far, giving a smooth, simple decision boundary.
* High gamma means the influence is close, making a complex, wiggly boundary that can overfit.
* Gamma is like the inverse of the radius of influence around data points.
* Choosing the right gamma helps balance between underfitting (too smooth) and overfitting (too complex).
* Usually, intermediate gamma values give the best model performance.
* Gamma works alongside the C parameter, which decides tolerance to misclassification.

## 12. What is the Naïve Bayes classifier, and why is it called "Naïve"?

Answer:  

Naive Bayes is a simple and fast classification algorithm based on Bayes Theorem.It predicts the class of an input by calculating the probability that the input belongs to each class, based on the features given.

**Working**

* Calculate prior probability of each class (how common each class is).

* Calculate likelihood: probability of features given each class.

* Use Bayes' Theorem to calculate posterior probability for each class.

* Predict the class with the highest posterior probability.

P(class/x) = p(x/class) * p(class) / p(x)

p(x/class) = likelihood

p(class) = class prior probality

p(x)= predictor prior probality

P(class/x) = Posterior probality

**Why it is called Naive?**

- It assumes all features are independent of each other given the class. This assumption is called naive because, in real life, features often depend on each other.

- This simplification makes the calculation much easier and faster, even if the independence assumption is not fully true.

Real world application :  

1. Recommendation Systems: Supports product recommendations by predicting user preferences based on their purchase history.

2. Fraud Detection: Identifies fraudulent transactions by modeling normal vs. abnormal behavior in financial data.

3. Document Classification: Categorizes news articles or documents into topics like sports, politics, or technology.

4. Spam Detection: Used to classify emails as spam or non-spam based on word patterns like win money or free prize.

5. Sentiment Analysis: Analyzes customer reviews or social media posts to classify sentiments as positive, negative, or neutral.





## 13. What is Bayes Theorem?

Answer :  

Bayes Theorem helps calculate the probability of an event happening given new evidence. It updates the prior belief with new data.In machine learning, it is used to calculate the likelihood that a hypothesis or prediction is true given the observed data.

- It combines prior knowledge (what you believed before) with new data (evidence) to give an updated probability (posterior).

P(A/B) = P(B/A) * P(A) / P(B)

1. P(A/B)= Probability of event A given B (posterior)
2. P(B/A)= Probability of event B given A (likelihood)
3. P(A)= Prior probability of A
4. P(B)= Probability of B

- It is fundamental in probabilistic models and classifiers like Naive Bayes.
- Enables decision making under uncertainty by updating beliefs when new evidence arrives.
- helps machines learn from data by continuously refining the chances of outcomes based on evidence.

**Example :**

1. Suppose 1% of people have an allergy.
2. A test detects the allergy correctly 80% of the time (true positive).
3. The test gives a false positive 10% of the time for people without allergy.
4. If someone tests positive, what is the chance they really have the allergy?

prior_probality = P(allergy) = 0.01

Probability test says yes if allergic=p(yes/allergic) = 0.8

Probability test says yes if not allergic: p(no/allergic) = 0.1

Overall probability test says yes:

              p(+ve)=P(+ve/disease) * p(disease)+ p(+ve/no disease) * P(no disease)

              p(yes) = (0.01 * 0.8)+ (0.99* 0.1) = 0.107

using bayes therom :                 

                  P(allergy/yes) = 0.8 * 0.01 /0.107 = 0.0748


So, even with a positive test, only about 7.5% chance the person actually has the allergy because false positives happen.

## 14. Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes?

Used when features are continuous and assumed to follow a normal (Gaussian) distribution. It calculates the probability of a feature value based on the Gaussian formula.

## 16. What are the key assumptions made by Naïve Bayes?

Answer :

1. Conditional Independence:
      All input features are assumed to be independent of each other given the class label. This means the presence or value of one feature does not affect others.This simplifies calculations a lot, even though it's often not true in real life

2. Equal Importance:
      All features contribute equally and independently to the outcome. The model treats every feature as equally important in predicting the class.

3. Sufficient Training Data:
      model needs enough examples for each class and feature combination to accurately estimate the chance of seeing certain features in that class. If there isnt enough data, the probability estimates can be unreliable or zero, making the model less accurate. Enough data helps the model learn the true patterns of how features appear in each class so it can make good predictions.

4. No Redundancy:
      features should give different information about the class without overlapping or repeating the same facts. If two features tell you the same thing, the model can't learn anything extra from having both, and it may overcount that information, which can reduce accuracy. Ideally, each feature should provide unique insight about the class to help the model make better predictions.

5. Balanced Classes:
      The training data ideally has a balanced distribution of classes to avoid bias toward majority classes.

## 20. How does Laplace Smoothing help in Naïve Bayes?

Answer :

Laplace Smoothing helps Naïve Bayes by solving the zero probability problem.

- Sometimes, a word or feature may not appear in the training data for a certain class, causing the model to assign it a zero probability.
- Multiplying probabilities causes the entire prediction to become zero, which is wrong.
- Laplace Smoothing adds a small positive number (usually 1) to all feature counts.
- This ensures that no probability is ever zero, so the model can handle unseen words or features better.
- It improves the model's ability to generalize and make reliable predictions on new data.

Example : If the word free never appeared in non-spam emails in training, without smoothing its probability would be zero for non-spam. With Laplace Smoothing, the probability becomes small but not zero, helping correct classification.

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', label='Class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('make_circles Dataset Visualization')
plt.legend()
plt.show()
```

```python
# create model using polynomial kernal
model=SVC(kernel="poly",degree=5)
model.fit(X,y)
```

```python
# plotting decision boundray

def plot_decision_boundary(model, X, y):
    #create grid
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot decision boundary and margins
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.viridis)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.viridis, s=40, edgecolors='k')
    plt.title("SVM with Polynomial Kernel (degree 5)")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```

```python
plot_decision_boundary(model, X, y)
```

## 27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually?

Answer :

```python
# genrate data using make_blob
from sklearn import datasets
X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=6)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', label='Class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('make_circles Dataset Visualization')
plt.legend()
plt.show()
```

```python
# genrate data using make_blob
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt

# Create a mesh grid for plotting decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

plt.figure(figsize=(15, 12))

# Re-initialize C as a list to prevent the TypeError
C_values = [0.1, 0.5, 1.0, 5.0]

for i, C in enumerate(C_values, 1):
    # Train SVM with linear kernel and different C
    svm = SVC(kernel='linear', C=C)
    svm.fit(X, y)

    # Predict on mesh grid points
    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot decision boundary and margin
    plt.subplot(2, 2, i)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.title(f'SVM Decision Boundary with C={C}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())

    # Plot support vectors
    sv = svm.support_vectors_
    plt.scatter(sv[:, 0], sv[:, 1], s=100, facecolors='none', edgecolors='k', label='Support Vectors')
    plt.legend()

plt.tight_layout()
plt.show()
```

## 45. Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC score.

Answer :

```python
from sklearn.datasets import load_digits
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import label_binarize
```

```python
digits = load_digits()
X = digits.data
y = digits.target
```

```python
# Binarize the output for multiclass ROC-AUC calculation
y_bin = label_binarize(y, classes=digits.target_names)
n_classes = y_bin.shape[1]
```

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train Gaussian Naive Bayes classifier
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
# Predict class probabilities on the test set
y_score = gnb.predict_proba(X_test)
```

```python
# Binarize test labels for ROC-AUC calculation
y_test_bin = label_binarize(y_test, classes=digits.target_names)
```

```python
# Calculate ROC-AUC score (one-vs-rest, macro average)
roc_auc = roc_auc_score(y_test_bin, y_score, average='macro', multi_class='ovr')
```

```python
print(f"ROC-AUC score (macro-average): {roc_auc:.3f}")
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns




# Plot ROC curves for each class
plt.figure(figsize=(10, 7))
colors = plt.cm.Set1(np.linspace(0, 1, n_classes))

for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc = roc_auc_score(y_test_bin[:, i], y_score[:, i])
    plt.plot(fpr, tpr, color=colors[i], lw=2,
             label=f'Class {i} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curves - Naive Bayes (Digits Dataset)')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
#The ROC-AUC score of 0.983 is excellent, meaning the model can very well distinguish between classes. It measures how well the classifier ranks
#positive instances higher than negative ones, so a score close to 1 shows strong ability to separate classes.
#The ROC curves plot this performance for each digit class by showing the balance between true positives and false positives at various thresholds.
# Steep curves going toward the top-left corner mean the model achieves high accuracy without many false alarms. High AUC values per class
# (around or above 0.95) confirm consistent good performance across digits.
```
