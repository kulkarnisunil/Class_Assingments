## 1.  What is Simple Linear Regression?

Answer :   

Simple Linear Regression is a statistical and supervised machine learning method used to model the linear relationship between two variables — one independent (predictor) variable and one dependent (target) variable. It fits a straight line that best predicts the target variable from the input variable.

y = mx+b

y is predicted variable or dependent variable.

X is independent variable

m is slope or coeffcient : how much y changes with each unit change in x.

b is intercept : value of y when x=0

The goal is to find the best-fitting line that minimizes the difference between actual and predicted values (often minimizing the sum of squared errors). Simple linear regression is widely used for prediction and understanding the linear effect of one variable on another when the relationship is approximately linear

Example :
1. calcualting selling price of card based on age of car
2. number of study hourse and marks attain by student

## 2. What are the key assumptions of Simple Linear Regression?

Answer :    

**Assumptions of Simple Linear Regression:**

1. Linearity:  The Relationship between dependent variable(y) and independent variable is linear.This means a change in X leads to a proportional change in Y.

2. Independence of Errors: the residuals (the differences between the observed and predicted values) in a regression model are not related or correlated with each other. means the error  get for one observation should not give any clue or predict the error you get for another observation.No autocorrelation exists, especially important in time-series data.

3. Homoscedasticity: the spread of errors or differences between actual and predicted values stays roughly the same no matter the value of X.ensure that the regression model estimates are reliable, efficient, and hypothesis tests valid and good model.

4. Normality of Residuals: The residuals are normally distributed, which allows for valid hypothesis testing and confidence intervals.

5. No Multicollinearity:Independent variables should not be highly correlated with each other(More relevant for multiple regression).




## 3.  What is heteroscedasticity, and why is it important to address in regression models?

Answer :   

Heteroscedasticity means that the spread of errors (residuals) in a regression model is not constant—it changes as the values of the predictor change. means input is small the errors might be small, but for larger input values, the errors become bigger.

**why is it important to address in regression models?**

- regression models assume errors have constant spread (homoscedasticity). When heteroscedasticity is present, it can mess up the accuracy of standard errors, confidence intervals, and hypothesis tests. This can lead to wrong conclusions about which variables really matter.

Its usefull beacuse  it shows how multiple factors together influence the outcome, helping make better, accurate predictions. This is widely used in real-world problems where many things affect the result.

## 5. What is polynomial regression, and how does it differ from linear regression?

Answer :   

Polynomial regression is an extension of linear regression used to model the relationship between a dependent variable and one or more independent variables when the relationship is non-linear. While linear regression fits a straight-line equation y=mx+b  polynomial regression fits an n-th degree polynomial equation:

formula :   y = b + m_1*x + m_2*x^2 + m_3 * x^3 .........

- Polynomial regression models data with bending curves, not just straight lines.
- It's useful when data shows a clear curve shape.
- The degree (power of x) controls how flexible the curve is.
- Too high degree can overfit the data

- Example :  polynomial regression is modeling the relationship between years of experience and salary in a job. In many cases, salary growth is not linear — after a few years, the salary increases faster due to promotions or skill mastery. Polynomial regression helps capture this curved relationship better than simple linear regression.

- widely used in finance (modeling stock trends), healthcare (disease progression, tissue growth), transportation (speed regulation), and public health (mortality rate estimation) to model complex, real-world relationships that are non-linear

**Difference:**

1. Linear regression fits straight lines; polynomial regression fits curves.
2. Linear regression is simpler, easier to interpret; polynomial is more complex and flexible.
3. Linear regression works for linear relationships; polynomial is for non-linear data.
4. Polynomial regression may overfit if degree is too high; linear regression is less prone to this.


## 6. Implement a Python program to fit a Simple Linear Regression model to the following sample data:
## ● X = [1, 2, 3, 4, 5]
## ● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
## Plot the regression line over the data points.

Answer :   

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X=np.array([1,2,3,4,5]).reshape(-1,1) # linear regression except the input features to be a 2-dimensional array,row and clmn
X.shape
```

```python
Y= np.array([2.1, 4.3, 6.1, 7.9, 10.2])
# X = dependent and y = indepedent
```

```python
model = LinearRegression()
```

```python
# fit model on data
model.fit(X,Y)
```

```python
# Predict values using the model
Y_pred = model.predict(X)
```

```python
Y_pred
```

```python
# Plot data points
plt.scatter(X, Y, color='blue', label='Data points')
# Plot regression line
plt.plot(X, Y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(Y,Y_pred)
# difference between actual and predict point is 0.0167
```

```python
data
```

```python
X
```

```python
print(f"Intercept : {model.intercept_}")
print(f"Coefficient : {model.coef_}")
```

```python
# Add constant term for VIF calculation
X['Intercept'] = 1
```

```python
# Calculate VIF for each feature
vif_df = pd.DataFrame()
vif_df['feature'] = X.columns
vif_df['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
```

```python
print(vif_df)
```

```python
# If VIF > 5 (sometimes 10), it indicates high multicollinearity between features.
# VIF Value for area and rooms are above 5 ,suggesting that there is some multicollinearity between these two features. This means that
# area and rooms are somewhat linearly related,which can make it harder for the model to isolate the individual impact of each variable on price.
# High multicollinearity can make the model unstable and coefficients unreliable.
# The intercept typically has a very high VIF which can be ignored.
# consider removing or combining correlated features if VIF indicates a problem.
```

## 9. Question 9: Create a residuals plot for a regression model trained on this data:
## ● X = [10, 20, 30, 40, 50]
## ● Y = [15, 35, 40, 50, 65]
## Assess heteroscedasticity by examining the spread of residuals.

Answer :   

**Assessing Heteroscedasticity:**


* If residuals are randomly scattered with constant spread, heteroscedasticity is unlikely.

* If residual spread increases or decreases with X (forming a funnel shape), heteroscedasticity may be present, indicating non-constant error variance.

```python
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])
```

```python
model = LinearRegression()
```

```python
model.fit(X, Y)
```

```python
y_pred = model.predict(X)
```

```python
y_pred
```

```python
# Calculate residuals
residuals = Y - Y_pred
```

```python
residuals
```

```python
# Plot residuals vs X
plt.scatter(X, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.title('Residuals Plot')

plt.show()
```

```python
# The blue points (residuals) scatter around the red dashed line at zero, which means the model predictions are close to actual values.
# There is no clear increasing or decreasing spread of residuals, suggesting the error variance is roughly constant (homoscedasticity).
# Points are close to the zero line; no big outliers are seen.
# No obvious pattern or curve in the residuals means the linear model fits well.

# The residuals plot shows no clear pattern of increasing or decreasing spread as X increases (no funnel or cone shape). The residuals have roughly
# equal variance at all levels of X, suggesting constant variance (homoscedasticity).In this data, no heteroscedasticity or non linearity.
```

## 10.  Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.

Answer :

### heteroscedasticity :

Heteroscedasticity means the variance of the errors (residuals) in a regression model is not constant across all levels of the independent variables. In simpler terms, the spread of prediction errors changes with the size of the input variables.

- Heteroscedasticity leads to inefficient and biased standard errors impacting inference.
- It violates a key assumption of ordinary least squares (OLS) regression.
-auses unreliable standard errors, confidence intervals, and hypothesis tests.
- Ultimately, it can give misleading conclusions about the impact of predictors.

Algorithm influenced by heteroscedasticity:

- Linear regression and other parametric regressions are affected.
- Standard errors and p-values become unreliable.
- Predictions might remain unbiased but less precise.

Algirthm not influced by heteroscedasticity:
- Tree-based models like Random Forest, Gradient Boosting.
- Non-parametric models like K-Nearest Neighbors, SVM.
- Neural networks.

**Steps to address Heteroscedasticity:**

1. Transform the dependent variable (e.g., apply a log transform to house prices) to stabilize variance.
2. Use weighted least squares regression, giving less weight to data points with higher variance.this stabilizing the impact of outliers and variable errors.
3. Redefine the dependent variable by using rates or ratios instead of raw values if appropriate.for example, For example, predicting price per square foot or price per room can reduce inherent variability.
4. robust regression methods that are less sensitive to violations of constant variance, such as quantile regression or Theil-Sen estimators.
5. Check residual plots again after adjustments to ensure constant variance.

### Multicollinearity :

Multicollinearity means that two or more features (independent variables) in regression model are highly correlated with each other. This causes problems because it becomes hard to tell how each feature individually impacts the target (like house price), and the estimated effects can be unstable or misleading and reduce model interpretability.

Algorithm influenced by Multicollinearity:

1. Linear Regression: Strongly affected because correlated predictors inflate coefficient variance and make estimates unstable.

2. Logistic Regression: Also affected since its a form of regression, causing unstable odds ratio estimates.

3. Generalized Linear Models (GLMs): Similarly sensitive as their parameters are estimated via regression techniques.

4. Stepwise Regression: Vulnerable due to reliance on coefficient significance.


Algorithm not influenced by Multicollinearity:

1. Tree base model:  They split on features independently, so multicollinearity usually doesn't harm performance or coefficient interpretation.[decision tree,radnom forest,bagging and boosting]

2. Support Vector Machines (SVMs) and Neural Networks: They focus on decision boundaries rather than coefficient estimates, so multicollinearity is less critical.

3. K-Nearest Neighbors (KNN), Naive Bayes: Not affected by coefficient interpretation issues.


**Steps to address Multicollinearity:**

1. Calculate Variance Inflation Factor (VIF) for the features to detect highly correlated predictors.Drop one of the features with a high VIF if it does not harm model performance significantly.

2. Remove or combine correlated features, such as merging similar variables or using domain knowledge to retain the most relevant ones.

3. Apply dimensionality reduction techniques like Principal Component Analysis (PCA) if many features are correlated.[combine correlated features into a single uncorrelated component—although this can affect interpretability]

4. Use regularization methods like Ridge or Lasso regression, which penalize large coefficients and reduce multicollinearity effects.

**Steps for Ensuring a Robust Model:**       

1.
    - start by visualizing data with scatterplots and residual plots to spot patterns, outliers, and unequal error spread.
    - calculate VIF to detect if any predictors are highly correlated, which can affect model reliability.
    - This helps us find problems early and understand data's structure before modeling.

2.  Apply cross-validation to assess the generalizability of the model and to ensure that mitigation strategies (like regularization or transformation) improve out-of-sample predictive accuracy.

3. compare performance before and after addressing these issues to ensure better and stable predictions.or use alternative algorithm such as ensemble methods or robust regressors

**In this case,predicting house prices with features like area, rooms, and location—start with exploratory data analysis to find signs of heteroscedasticity (residual plots vs. fitted values) and multicollinearity (correlation matrix, VIF). Then, follow with log transformation of prices, check VIF for predictors, and consider ridge regression for both predictive power and interpretability. Carry out these steps iteratively, validating and tuning at each stage to produce a model ready for production deployment**
