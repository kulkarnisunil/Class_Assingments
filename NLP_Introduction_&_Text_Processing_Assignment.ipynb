#Question 1: What is Computational Linguistics and how does it relate to NLP?

Answer :

Computational Linguistics is the study of human language using computational methods and algorithms. It combines ideas from linguistics and computer science to help machines understand and process natural language.

Natural Language Processing (NLP) is closely related to computational linguistics because it applies these theoretical concepts to build practical applications such as chatbots, translation systems, and sentiment analysis tools. In simple terms, computational linguistics focuses on the theory, while NLP focuses on practical implementation.

# Question 2: Briefly describe the historical evolution of Natural Language Processing.

Answer :

The development of NLP has gone through several stages. Early systems in the 1950s and 1960s were rule-based and relied on manually written grammar rules. Later, in the 1990s, statistical methods became popular, where computers learned patterns from large text datasets. With the advancement of machine learning, NLP models improved significantly. Today, modern NLP uses deep learning techniques such as neural networks and transformer models to achieve highly accurate language understanding.

# Question 4: What is text normalization and why is it essential in text processing tasks?

Answer:

Text normalization is the process of converting text into a standard and consistent format before applying NLP techniques. It may include converting text to lowercase, removing punctuation, correcting spelling, and reducing words to their base form.

This step is important because raw text data often contains noise and inconsistencies. Normalization helps improve the quality of the data and makes NLP models more accurate and reliable.

# Question 5: Compare and contrast stemming and lemmatization with suitable examples.

Answer:

Stemming and lemmatization are techniques used to reduce words to their root form during text preprocessing.

Stemming removes suffixes from words using simple rules. For example, running may become runn.

Lemmatization converts words to their meaningful base form using vocabulary and grammar rules. For example, running becomes run and better becomes good.

Stemming is faster but less accurate, while lemmatization provides more meaningful results.

# Question 10: Imagine you are a data scientist at a fintech startup. You’ve been taskedwith analyzing customer feedback. Outline the steps you would take to clean, process,and extract useful insights using NLP techniques from thousands of customer reviews.

Answer:


1. Data Collection

First, collect customer feedback from different sources such as mobile app reviews, customer support tickets, surveys, and social media platforms. The data should be stored in a structured format like a CSV file containing fields such as review_id, date, review_text, and rating.

2. Text Cleaning and Preprocessing

The raw text data usually contains noise, so it must be cleaned before analysis. This includes removing HTML tags, URLs, and non-ASCII characters. The text is converted to lowercase and punctuation or unnecessary whitespace is removed. Contractions such as “don’t” are expanded to “do not”. In some cases, stop words may also be removed if they do not add meaningful information.

3. Tokenization and Normalization

Next, the text is split into individual words using tokenization. After that, normalization techniques such as lemmatization are applied to reduce words to their base form. For example, “running” becomes “run”. Libraries like NLTK or spaCy can be used for this step.

4. Exploratory Data Analysis

Once the text is processed, exploratory analysis can be performed. Word frequency analysis and word clouds help identify commonly used words in reviews. Additionally, n-grams can reveal frequent phrases. It is also useful to analyze the relationship between review length and ratings.

5. Sentiment Analysis

Sentiment analysis helps determine whether a review expresses a positive, negative, or neutral opinion. Pre-trained tools such as VADER or TextBlob can be used to assign sentiment scores. Aggregating these scores across time or product categories can reveal customer satisfaction trends.

6. Topic Modeling

Topic modeling techniques such as Latent Dirichlet Allocation (LDA) can identify hidden themes in the reviews. For example, topics might represent issues related to customer support, mobile application performance, or transaction fees. These topics can then be interpreted and labeled for better understanding.

7. Entity and Aspect Extraction

Named Entity Recognition (NER) can be used to detect important entities such as product names, companies, or services mentioned in reviews. Aspect-based sentiment analysis can further identify how customers feel about specific features like interest rates, payment speed, or user interface.

8. Insights and Reporting

Finally, the extracted insights are summarized and presented to business teams. This may include identifying the most common complaints, areas where customers are satisfied, and emerging problems. A dashboard or report can help stakeholders monitor customer sentiment and improve the product continuously.


```python
import pandas as pd
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
data = {
    "review": [
        "The mobile app keeps freezing and transactions fail sometimes.",
        "Customer service was helpful and solved my problem quickly.",
        "The interface is simple and easy to navigate.",
        "Account verification took too long and was frustrating.",
        "I like the new payment feature, it works smoothly."
    ]
}
```

```python
df = pd.DataFrame(data)
df.head()
```

#

```python
# Text preprocessing
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def process_review(text):
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))

    tokens = word_tokenize(text)

    clean_tokens = []
    for word in tokens:
        if word not in stop_words:
            base = lemmatizer.lemmatize(word)
            clean_tokens.append(base)

    return " ".join(clean_tokens)
```

```python
df["processed"] = df["review"].apply(process_review)
```

```python
# Sentiment score calculation
def sentiment_score(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

df["sentiment"] = df["processed"].apply(sentiment_score)
```

```python
# Topic extraction using TF-IDF + NMF
vectorizer = TfidfVectorizer(max_features=50)
matrix = vectorizer.fit_transform(df["processed"])
```

```python
topic_model = NMF(n_components=2, random_state=10)
topic_model.fit(matrix)
```

```python
words = vectorizer.get_feature_names_out()
```

```python
print("Discovered Topics:\n")

for i, topic in enumerate(topic_model.components_):
    top = topic.argsort()[-5:]
    topic_words = [words[j] for j in top]
    print(f"Topic {i+1}:", ", ".join(topic_words))
```

```python
print("\nAverage Sentiment Score:", df["sentiment"].mean())
print("\nMost Positive Review:")
print(df.loc[df["sentiment"].idxmax(), "review"])

print("\nMost Negative Review:")
print(df.loc[df["sentiment"].idxmin(), "review"])
```
